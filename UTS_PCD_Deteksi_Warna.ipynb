```python
#IMPORT LIBRARY
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('bahan-uji-warna1.jpg')
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
#Biru
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([140, 255, 255])

#Hijau
lower_green = np.array([40, 50, 50])
upper_green = np.array([80, 255, 255])

#Merah
lower_red = np.array([160, 50, 50])
upper_red = np.array([180, 255, 255])
```

```python
mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_green = cv2.inRange(hsv, lower_green, upper_green)
mask_red = cv2.inRange(hsv, lower_red, upper_red)
```

```python
# Deteksi Biru
blue_detected = cv2.bitwise_and(img, img, mask=mask_blue)

# Deteksi Hijau
green_detected = cv2.bitwise_and(img, img, mask=mask_green)

# Deteksi Merah
red_detected = cv2.bitwise_and(img, img, mask=mask_red)

plt.imshow(blue_detected)
plt.show()

plt.imshow(green_detected)
plt.show()

plt.imshow(red_detected)
plt.show()
```

```python
cv2.imwrite('biru_terdeteksi.jpg', blue_detected)
cv2.imwrite('hijau_terdeteksi.jpg', green_detected)
cv2.imwrite('merah_terdeteksi.jpg', red_detected)
```

```python
plt.subplot(2,2,3)
plt.imshow(mask_blue, cmap='gray')
plt.title('Deteksi Biru')

plt.subplot(2,2,4)
plt.hist(mask_blue.ravel(), 256, [0,256])
plt.title('Histogram Deteksi Biru')

plt.show()

plt.subplot(2,2,3)
plt.imshow(mask_green, cmap='gray')
plt.title('Deteksi Hijau')

plt.subplot(2,2,4)
plt.hist(mask_green.ravel(), 256, [0,256])
plt.title('Histogram Deteksi Hijau')

plt.show()

plt.subplot(2,2,3)
plt.imshow(mask_red, cmap='gray')
plt.title('Deteksi Merah')

plt.subplot(2,2,4)
plt.hist(mask_red.ravel(), 256, [0,256])
plt.title('Histogram Deteksi Merah')

plt.show()
```
